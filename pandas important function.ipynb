```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data={'age':[25,22,18,30,45,50,35,20,55,40],
         'gender': ['M','F','F','M','M','F','M','F','M','M'],
        'SCORE':[90,50,75,95,70,85,75,90,95,85]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

1.Value_counts()

```python
df['gender'].value_counts()
```

```python
df['gender'].value_counts(normalize=True)
#relative frequency of each value
```

2.Where

```python
df.where(df['age']>30)
```

```python
df.where(df['age']>30,other=0)
```

```python
df[df.where(df['age']>30,other=0).all(1)]
```

```python
con1=df['age']>30
con2=df['gender']=='F'
```

```python
df[df.where(con1 & con2,other=0).all(1)]
```

3.isin()

```python
df[df['age'].isin([30,20])]
```

```python
df[['age','gender']].isin({'age':[25,45],'gender':['M']}).all(1)
```

4.cut and qcut


```python
#it is used divide continuous variable into descrete bins
```

```python
df
```

```python
df['score bins']=pd.cut(df['SCORE'],bins=[40,50,60,70,80,90,100])
```

```python
df
```

```python
label_names=['very low','low','medium','high','very high','extreme']
```

```python
df['score bins']=pd.cut(df['SCORE'],bins=[40,50,60,70,80,90,100],labels=label_names)
```

```python
df
```

Qcut()

qcut is used to divide continuous variable into discrete bins based on quartiles

```python
df['age bins']=pd.qcut(df['age'], 4,labels =['young','mid young','mid adult','adult'])
```
