# Discount Analysis:

### What percentage of customers typically uses discounts?


```python
df['Discount Applied'].replace(True,'Discounts',inplace=True)
```

```python
df['Discount Applied'].replace(False,'No Discounts',inplace=True)
```

```python
status_d = df['Discount Applied'].value_counts()
key_list = list(status_d.keys())
value_list = status_d.values
fig1, ax1 = plt.subplots()
ax1.pie(value_list, labels=key_list, autopct='%1.1f%%', shadow=False, startangle=90)
ax1.axis('equal')
plt.title('Comparison of Total Spend: Discounts vs. No Discounts')
plt.show()
```

### How does the application of discounts correlate with the average rating given by customers?


```python
fig,axs =plt.subplots(1,2,figsize=(10,5))
sns.boxplot(x=df['Discount Applied'],y=df['Average Rating'],ax=axs[0])
axs[0].set_title('Discpunt applied vs Avereage Rating')

sns.barplot(x=df['Discount Applied'],y=df['Average Rating'],ax=axs[1])
axs[1].set_title('Discpunt applied vs Avereage Rating')

plt.tight_layout()
plt.show()
```

### Is there a significant difference in total spend between customers who use discounts and those who don't?

```python
sns.boxplot(x=df['Discount Applied'],y=df['Total Spend'])

plt.show()
```
