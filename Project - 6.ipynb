7. Find Out Number of Courses Per Subjects


```python
df.columns
```

```python
df['subject'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' contains your DataFrame with a 'subject' column
sns.countplot(data=df, x='subject')

# Optionally, you can set labels and customize the plot
plt.xlabel('Subject')
plt.ylabel('Count')
plt.title('Count Plot of Subjects')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

13. Find Most Popular Course Title


```python
df[df['num_subscribers'].max()==df['num_subscribers']]['course_title']
```

14. Display 10 Most Popular Courses As Per Number of Subscribers


```python
x=df.sort_values('num_subscribers',ascending=False).head(10)
```

```python

plt.xticks(rotation=90)
sns.barplot(y='course_title',x='num_subscribers',data=x)
```

16. Does Price Affect the Number of Reviews?

```python
plt.figure(figsize=(15,3))
sns.scatterplot(x='price',y='num_reviews',data=df)
plt.legend(['price'])
```

17. Find Total Number of Courses Related To Python


```python
python_coures=df[df['course_title'].str.contains('python',case=False)]
len(python_coures)
```

18. Display 10 Most Popular Python Courses As Per Number of Subscribers


```python
python_10=python_coures.sort_values(by='num_subscribers',ascending=False).head(10)
```

```python

plt.xticks(rotation=90)
sns.barplot(y='course_title',x='num_subscribers',data=python_10)
```
