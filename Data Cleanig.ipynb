```python
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
os.listdir('./data_sets')
```

## Reading Dataset

```python
df = pd.read_excel("./data_sets/Customer Call List.xlsx")
```

```python
df
```

## Droping Duplicates

```python
df.drop_duplicates(inplace=True)
df
```

## Removing Column

```python
df.drop(columns="Not_Useful_Column",inplace=True)
df
```

## Removing unwanted letters 

```python
df["Last_Name"]=df["Last_Name"].str.strip("123._/")
```

```python
df
```

## Removing Symbols

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"]=df['Phone_Number'].apply(lambda x:str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].str.strip("na")
df
```

## Removing nan i.e(not a number)

```python
df["Phone_Number"] = df["Phone_Number"].str.strip("NaN")
```

```python
df
```

```python
df[['street_address', 'zipcode', 'area']] = df['Address'].str.split(',', n=2, expand=True)
```

```python
df.sample(5)
```

```python
df["Paying Customer"]=df['Paying Customer'].str.replace("No",'N')
```

```python
df["Paying Customer"]=df['Paying Customer'].str.replace("Yes",'Y')
```

```python
df["Do_Not_Contact"]=df['Do_Not_Contact'].str.replace("No",'N')
df["Do_Not_Contact"]=df['Do_Not_Contact'].str.replace("Yes",'Y')
```

## Filling none with blank Space

```python
df=df.fillna('')
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
```
