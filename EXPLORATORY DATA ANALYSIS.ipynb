## 5) covert data into matrix

```python

non_numeric_columns = df.select_dtypes(exclude=['number']).columns


for column in non_numeric_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')


correlation_matrix = df.corr()
df1=correlation_matrix
```

```python
df2 =df.pivot(columns=['Rank', 'CCA3', 'Country'])
```

## 7) plot heat map using sns ,increse size using plt.fig 

```python
plt.title("world population")
sns.heatmap(df1,annot=True);
```

## 8) find the mean of every continents and sort columns in decreseing order

```python
df3=df.groupby("Continent")[['1970 Pop',
       '1980 Pop', '1990 Pop', '2000 Pop',
       '2010 Pop', '2015 Pop', '2020 Pop',
       '2022 Pop']].mean().sort_values(by ="2022 Pop",ascending =False)
```

```python
df.rename(columns={'2020 Population': '2020 Pop'}, inplace=True)
df.rename(columns={'2022 Population': '2022 Pop'}, inplace=True)
df.rename(columns={'2015 Population': '2015 Pop'}, inplace=True)
df.rename(columns={'2010 Population': '2010 Pop'}, inplace=True)
df.rename(columns={'2000 Population': '2000 Pop'}, inplace=True)
df.rename(columns={'1990 Population': '1990 Pop'}, inplace=True)
df.rename(columns={'1980 Population': '1980 Pop'}, inplace=True)
df.rename(columns={'1970 Population': '1970 Pop'}, inplace=True)
```

```python
plt.figure(figsize=(10,10))
import matplotlib.pyplot as plt


plt.rcParams.update({'font.size': 5})
df3.transpose().plot();
```

```python
import matplotlib.pyplot as plt


plt.rcParams.update({'font.size': 5})

df3.boxplot()

plt.show()
```
