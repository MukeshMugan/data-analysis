### Hypothesis 1: Technology products have the highest profit margin compared to other product categories.

```python
df.columns
```

```python
plt.title("profit based on the categories")
plt.ylabel('profit')
df.groupby('Category')['Profit'].sum().plot(kind='bar',color=['red','blue','green'])
```

### Hypothesis 2: The East region has the highest sales compared to other regions.

```python
df.columns
```

```python
df.pivot_table(index='Region',values='Sales',aggfunc='sum').plot(kind='bar')
```

### Hypothesis 3: Sales are higher during certain months of the year.

```python
df.columns
```

```python
df['Order Month']=df['Order Date'].dt.month
```

```python
plt.title('sales by month')
plt.ylabel
df.groupby('Order Month')['Sales'].sum().plot(kind='line')
```

### Hypothesis 4: Orders with same-day shipping have the lowest rate of returned products.

```python
x=df.groupby('Ship Mode').size()
y=df[df['Profit']<=0].groupby('Ship Mode').size()
```

```python
z=(y/x)*100
```

```python
z.plot(kind='bar')
```

### Hypothesis 5: The company's profit is more on weekdays than on weekends.

```python
df['Day_of_Week'] = df['Order Date'].dt.day_name()
```

```python
df.groupby("Day_of_Week")['Profit'].sum().plot(kind='bar')
```
