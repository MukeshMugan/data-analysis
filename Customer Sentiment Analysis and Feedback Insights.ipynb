```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./data_sets/purchase behaviou dataset.csv')
df
```

```python
df['Couples'].fillna('Not couples',inplace =True)
```

```python
df['Couples'] =df['Couples'].str.title()
```

```python
df['LIFESTAGE']=df['LIFESTAGE'].str.title()
```

```python
df.isnull().sum()
```

```python
df['PREMIUM_CUSTOMER'].unique()
```

```python
sns.countplot(y=df['LIFESTAGE'])
plt.grid()
```

```python
df['LIFESTAGE'].value_counts()
```

```python
crosstabs= pd.crosstab(df['LIFESTAGE'],df['PREMIUM_CUSTOMER'])
```

```python
crosstabs.plot(kind='bar')
plt.grid()
plt.show()
```

```python
df['PREMIUM_CUSTOMER'].value_counts().plot(kind='bar',color=['Red','green','yellow'])
plt.grid()
plt.title("Types of Subscriber's count")
plt.show()
```
